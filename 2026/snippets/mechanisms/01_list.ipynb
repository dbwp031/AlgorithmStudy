# 배열
Python은 list라는 동적 배열 자료구조를 구현한 객체를 가지고 있다.

```python
lst = [1,2,3]
print(id(lst))
lst.append(4)
print(id(lst))
```

```python
import ctypes

lst = [1, 2, 3]

# CPython의 list 내부 구조체에 직접 접근
class PyListObject(ctypes.Structure):
    _fields_ = [
        ("ob_refcnt",  ctypes.c_ssize_t),
        ("ob_type",    ctypes.c_void_p),
        ("ob_size",    ctypes.c_ssize_t),   # len()과 동일
        ("ob_item",    ctypes.c_void_p),
        ("allocated",  ctypes.c_ssize_t),   # 실제 할당 용량
    ]

obj = PyListObject.from_address(id(lst))
print(f"ob_size   = {obj.ob_size}")    # 3  ← len(lst)
print(f"allocated = {obj.allocated}")  # 3 or 4

print(f"lst id: {id(lst)}")

lst.append(4)
print(f"lst id: {id(lst)}")

lst.append(5)
print(f"lst id: {id(lst)}")

obj = PyListObject.from_address(id(lst))
print(f"ob_size   = {obj.ob_size}")    # 3  ← len(lst)
print(f"allocated = {obj.allocated}")  # 3 or 4
```
